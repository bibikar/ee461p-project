```python
import pandas as pd
import os
pd.options.display.max_columns = 999
```

```python
outdir = '/home/sameer/msd'
os.makedirs(outdir, exist_ok=True)
```

```python
features = ['track_id', 'duration', 'end_of_fade_in', 'key', 'key_confidence', 'loudness',
            'mode', 'mode_confidence', 'start_of_fade_out', 'tempo', 'time_signature',
            'time_signature_confidence', 'artist_familiarity', 'artist_hotttnesss']

root = '/mnt/snap/data'

for dir1 in os.listdir(root):
    for dir2 in os.listdir(os.path.join(root, dir1)):
        outfile = os.path.join(outdir, '{}{}.h5'.format(dir1, dir2))
        if os.path.isfile(outfile):
            print('Skipping generation of {} because it already exists.'.format(outfile))
            continue
        dataframes = []
        for dir3 in os.listdir(os.path.join(root, dir1, dir2)):
            path = os.path.join(root, dir1, dir2, dir3)
            print('Processing in', path)
            
            for f in os.listdir(path):
                with pd.HDFStore(os.path.join(path, f), 'r') as h:
                    series = []
                    for key in h.keys():
                        series.append(h.get(key))
                    x = pd.concat(series, axis=1)[features]
                    dataframes.append(x)
        pd.concat(dataframes).to_hdf(outfile, 'data')
        del dataframes
```

```python
df = pd.concat(dataframes).reset_index()
```
