```python
import pandas as pd
```

```python
songs = pd.read_csv('kaggle_songs.txt', delim_whitespace=True, header=None, names=['song_id', 'id'])
```

```python
triplets = pd.read_csv('kaggle_visible_evaluation_triplets.txt', delim_whitespace=True, header=None, names=['user_id', 'song_id', 'count'])
```

```python
songs_to_tracks = pd.read_csv('taste_profile_song_to_tracks.txt', delim_whitespace=True, header=None, names=['song_id', 'track_id'])
```

```python
users = pd.read_csv('kaggle_users.txt', delim_whitespace=True, header=None, names=['user_id'])
```

```python
df = pd.merge(triplets, songs_to_tracks, on='song_id').drop('song_id', axis=1)
```

```python
users_we_want = df.groupby('user_id').count()[['track_id']] > 20
```

```python
users_we_want = users_we_want[users_we_want['track_id']].reset_index()
```

```python
df.merge(users_we_want[['user_id']], on='user_id').dropna().to_hdf('diverse_listeners.h5', key='data')
```
